```python
plt.figure(figsize=(10, 6))
sns.regplot(data=word_count_stats, x='question_word_count', y='answer_word_count', scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('Correlation between Question and Answer Word Counts')
plt.xlabel('Question Word Count')
plt.ylabel('Answer Word Count')
plt.show()

correlation = word_count_stats['question_word_count'].corr(word_count_stats['answer_word_count'])
print(f"Pearson Correlation Coefficient: {correlation:.4f}")
```

The extremely low Pearson Correlation Coefficient (~0.051) suggests that the length of a medical question is a poor predictor of how long the answer will be. Here are a few reasons why:

Complexity of Content vs. Complexity of Query: A very simple, short question like "What is cancer?" requires a massive, detailed explanation. Conversely, a long, specific question about a very niche symptom might have a concise, direct answer.

Dataset Structure: Many questions in MedQuad follow a standardized, short template (e.g., "What is (are) [Condition]?" or "What are the treatments for [Condition]?"). While the questions are nearly identical in length, the answers vary wildly depending on whether the condition is a common disease or a rare genetic syndrome.

Source Variability: The data comes from different organizations (CDC, NIH, etc.). Some sources provide exhaustive summaries, while others provide brief bullet points, regardless of how the question was phrased.

Information Density: In medical contexts, clarity and completeness are prioritized over matching the "energy" or length of the user's inquiry.

```python
# PREPROCESSING FUNCTIONS

import re
import html
import unicodedata

def normalize_text(text: str, lowercase: bool = False) -> str:
    """
    Clean text while preserving readable natural language.
    """
    if pd.isna(text):
        return ""

    text = str(text)
    text = html.unescape(text)
    text = unicodedata.normalize("NFKC", text)
    text = re.sub(r"<[^>]+>", " ", text)
    text = text.replace("\r", " ").replace("\n", " ").replace("\t", " ")
    text = text.replace("–", "-").replace("—", "-")
    text = text.replace("“", '"').replace("”", '"')
    text = text.replace("‘", "'").replace("’", "'")
    text = re.sub(r"\s+", " ", text).strip()

    if lowercase:
        text = text.lower()

    return text


def is_low_quality_row(question: str, answer: str, min_answer_words: int = 6) -> bool:
    """
    Basic low-quality filter.
    """
    if not question or not answer:
        return True

    q_words = len(question.split())
    a_words = len(answer.split())

    if a_words < min_answer_words:
        return True

    if q_words < 2:
        return True

    bad_patterns = [
        r"^n/?a$",
        r"^unknown$",
        r"^not available$",
        r"^none$",
    ]
    answer_l = answer.lower().strip()
    if any(re.match(p, answer_l) for p in bad_patterns):
        return True

    return False


def preprocess_medquad(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # Keep original dataset source label
    df["source_name"] = df["source"]

    # Clean text
    df["question_clean"] = df["question"].apply(lambda x: normalize_text(x, lowercase=False))
    df["answer_clean"] = df["answer"].apply(lambda x: normalize_text(x, lowercase=False))

    # Remove rows with empty cleaned text
    df = df[(df["question_clean"] != "") & (df["answer_clean"] != "")].copy()

    # Remove obvious low-quality rows
    mask_low_quality = df.apply(
        lambda row: is_low_quality_row(row["question_clean"], row["answer_clean"]),
        axis=1
    )
    df = df[~mask_low_quality].copy()

    # Deduplicate using lowercase normalized versions
    df["_question_dedup"] = df["question"].apply(lambda x: normalize_text(x, lowercase=True))
    df["_answer_dedup"] = df["answer"].apply(lambda x: normalize_text(x, lowercase=True))
    df = df.drop_duplicates(subset=["_question_dedup", "_answer_dedup"]).copy()
    df = df.drop(columns=["_question_dedup", "_answer_dedup"])

    # Add stable document IDs AFTER dedup
    df = df.reset_index(drop=True)
    df["document_id"] = [f"doc_{i:06d}" for i in range(len(df))]

    # Make cleaned answer the default answer column for downstream chunking
    df["answer"] = df["answer_clean"]

    # Keep placeholder column for teammate compatibility
    df["source_page_url"] = ""

    df["content_question_answer"] = (
        "Question: " + df["question_clean"] + " Answer: " + df["answer_clean"]
    )

    df["question_word_count"] = df["question_clean"].str.split().str.len()
    df["answer_word_count"] = df["answer_clean"].str.split().str.len()

    return df


# ADVANCED PREPROCESSING

df_clean = preprocess_medquad(df)

print("Preprocessed shape:", df_clean.shape)
print(df_clean[[
    "document_id",
    "question_clean",
    "answer_clean",
    "source_name"
]].head())

df_clean.to_csv("medquad_preprocessed.csv", index=False)

print("Saved to medquad_preprocessed.csv")
print(f"Rows before preprocessing: {len(df)}")
print(f"Rows after preprocessing: {len(df_clean)}")
print(f"Rows removed: {len(df) - len(df_clean)}")
```
